### Q.2. What are the rules to create variables in Python?
**Answer:**
1. Variable names must start with a letter (a-z, A-Z) or an underscore (_).
2. The rest of the name can contain letters, numbers, or underscores.
3. Variable names are case-sensitive.
4. Reserved words (keywords) cannot be used as variable names.

### Q.3. What are the standards and conventions followed for the nomenclature of variables in Python to improve code readability and maintainability?
**Answer:**
1. **Snake Case:** Use lowercase letters separated by underscores (e.g., `variable_name`).
2. **Descriptive Names:** Use meaningful and descriptive names.
3. **Constants:** Use all uppercase letters separated by underscores (e.g., `CONSTANT_NAME`).
4. **Private Variables:** Start with a single underscore (e.g., `_private_var`).
5. **Avoid Single Letters:** Except for counters or iterators (e.g., `i`, `j`).

### Q.8. Explain the mutable and immutable data types with proper explanation & examples.
**Answer:**
- **Mutable Data Types:** These can be changed after creation. Examples include lists, sets, and dictionaries.
  ```python
  my_list = [1, 2, 3]
  my_list[0] = 4  # List is mutable
  ```
- **Immutable Data Types:** These cannot be changed after creation. Examples include strings, tuples, and integers.
  ```python
  my_tuple = (1, 2, 3)
  # my_tuple[0] = 4  # This will raise an error because tuples are immutable
  ```

### Q.9. Write a code to create the given structure using only for loop.
**Structure:**
```
height = 5
for i in range(height):
        # Print leading spaces
        for j in range(height - i - 1):
            print(" ", end="")
        # Print stars
        for k in range(2 * i + 1):
            print("*", end="")
        # Move to the next line
        print()
```

### Q.10. Write a code to create the given structure using while loop.
